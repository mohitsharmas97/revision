```python
# List of columns to keep, including 'diagnosis'
columns_to_keep = [
    'diagnosis',
    'concave points_worst',
    'concave points_mean',
    'radius_worst',
    'perimeter_worst',
    'area_worst',
    'concavity_mean',
    'concavity_worst',
    'radius_mean',
    'perimeter_mean',
    'area_mean'
]

# Create a new DataFrame with only the specified columns
dataset_filtered = dataset[columns_to_keep].copy()

# Display the head of the new DataFrame to confirm the columns
display(dataset_filtered.head())

# Display the shape of the new DataFrame
display(dataset_filtered.shape)
```

```python
input_data=dataset_filtered.drop('diagnosis',axis=1)
output_data=dataset_filtered['diagnosis'].map({'M': 1, 'B': 0})
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
input_data=pd.DataFrame(scaler.fit_transform(input_data), columns=input_data.columns)
```

```python
input_data.head(3)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
import tensorflow
from keras.layers import Dense
from keras.models import Sequential
```

```python
output_data.head(3)
```

```python
ann=Sequential()
```

```python
dataset_filtered.shape
```

```python
ann.add(Dense(116,input_dim=10,activation='relu'))
ann.add(Dense(units=8, activation='relu'))
# Add the output layer
ann.add(Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)
```

```python
# Evaluate the model on the training data
loss, accuracy = ann.evaluate(x_train, y_train, verbose=0)
print(f'Training Accuracy: {accuracy:.4f}')

# Evaluate the model on the testing data
loss, accuracy = ann.evaluate(x_test, y_test, verbose=0)
print(f'Testing Accuracy: {accuracy:.4f}')
```

```python
# Save the model in the native Keras format
ann.save('breast_cancer_model.keras')
print("Model saved to 'breast_cancer_model.keras'")
```
